# What is an outlier?

An outlier is a data point in a data set that is distant from all other observations. 
or
An outlier is a data point that lies outside the overall distribution of the dataset.

# Criterias to identify an outlier?
1. Using IQR 

Data point that falls outside of 1.5 times of an interquartile range above the 3rd quartile and below the 1st quartile

Data Value < Q1 - 1.5 IQR
Data Value > Q3 + 1.5 IQR

2. Z score

Data point that falls outside of 3 standard deviations. we can use a z score and if the z score falls outside of 2 standard deviation

z = (x-u/sd)
Any data above 3sd, are normally considered as outliers

# Detecting an outlier using Z score


Using Z score
Formula for Z score = (Observation — Mean)/Standard Deviation

z = (X — μ) / σ

```python
outliers=[]

def detect_outliers(data):
    
    threshold=3
    mean = np.mean(data)
    std =np.std(data)
    
    
    for i in data:
        z_score= (i - mean)/std 
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
outlier_pt=detect_outliers(dataset)
```
