Question 8: How do I apply multiple filter criteria to a pandas Dataframe

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[(movies.duration>=200) & (movies.genre =="Drama")]
```

```python
movies[(movies.duration>=200) | (movies.genre =="Drama")]
```

```python
movies[(movies.genre =="Drama") | (movies.genre =="Crime") | (movies.genre =="Action")]
```

```python
movies[movies.genre.isin(['Crime','Action','Drama'])]
```
